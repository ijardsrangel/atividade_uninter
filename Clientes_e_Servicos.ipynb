```python
# Parte 1 do Projeto: Desenvolvimento do Menu Principal e do Loop para escolha das opções
# Estruturas de dados para armazenar as informações
clientes = {}
servicos = {}
agendamentos = {}

# Função para exibir o Menu Principal
def mostrar_menu():
  print("\n### Gerenciamento de Clientes e Serviços ###")
  print("#### Autônomos e Profissionais Liberais ####")
  print("\n[1] Cadastrar Clientes")
  print("[2] Consultar Clientes")
  print("[3] Atualizar Clientes")
  print("[4] Remover Clientes")
  print("\n[5] Cadastrar Serviços")
  print("[6] Consultar Serviços")
  print("[7] Atualizar Serviços")
  print("[8] Remover Serviços")
  print("\n[9] Cadastrar Agendamento")
  print("[10] Consultar Agendamento")
  print("[11] Atualizar Agendamento")
  print("[12] Remover Agendamento")
  print("\n[0] Sair")
  escolha = input("Escolha uma opção (0-12): ")
  return escolha

# Loop principal do aplicativo
def main():
    while True:
        escolha = mostrar_menu()
        if escolha == '0':
            print("Saindo do aplicativo. Até logo!")
            break
        elif escolha == '1':
            cadastrar_cliente()
        elif escolha == '2':
            consultar_clientes()
        elif escolha == '3':
            atualizar_cliente()
        elif escolha == '4':
            remover_cliente()
        elif escolha == '5':
            cadastrar_servico()
        elif escolha == '6':
            consultar_servicos()
        elif escolha == '7':
            atualizar_servico()
        elif escolha == '8':
            remover_servico()
        elif escolha == '9':
            cadastrar_agendamento()
        elif escolha == '10':
            consultar_agendamentos()
        elif escolha == '11':
            atualizar_agendamento()
        elif escolha == '12':
            remover_agendamento()
        else:
            print("Opção inválida. Por favor, escolha uma opção válida.")

# Parte 2: Desenvolvimento das funções para cadastrar clientes, serviços e agendamentos
# Função para cadastrar clientes
def cadastrar_cliente():
  cpf = input("CPF do Cliente: ")
  if cpf in clientes:
      print(f'Erro: Cliente com CPF "{cpf}" já existe.')
      return
  nome = input("Nome: ")
  endereco = input("Endereço: ")
  telefone = input("Telefone: ")
  email = input("Email: ")
  cliente = {"Nome": nome, "Endereço": endereco, "Telefone": telefone, "Email": email}
  clientes[cpf] = cliente
  print(f'Cliente "{nome}" cadastrado com sucesso!')

# Função para cadastrar serviços
def cadastrar_servico():
  codigo = input("Código do Serviço: ")
  if codigo in servicos:
      print(f'Erro: Serviço com código "{codigo}" já existe.')
      return
  descricao = input("Descrição do Serviço: ")
  preco = input("Preço: ")
  servico = {"Descrição": descricao, "Preço": preco}
  servicos[codigo] = servico
  print(f'Serviço "{descricao}" cadastrado com sucesso!')

# Função para cadastrar agendamentos
def cadastrar_agendamento():
  cpf = input("CPF do Cliente: ")
  if cpf not in clientes:
      print(f'Erro: Cliente com CPF "{cpf}" não existe.')
      return
  cod_servico = input("Código do Serviço: ")
  if cod_servico not in servicos:
      print(f'Erro: Serviço com código "{cod_servico}" não existe.')
      return
  cod_agendamento = input("Código do Agendamento: ")
  if cod_agendamento in agendamentos:
      print(f'Erro: Agendamento com código "{cod_agendamento}" já existe.')
      return
  data = input("Data (dd/mm/yyyy): ")
  hora = input("Hora (HH:MM): ")
  agendamento = {"CPF Cliente": cpf, "Código Serviço": cod_servico, "Data": data, "Hora": hora}
  agendamentos[cod_agendamento] = agendamento
  print("Agendamento cadastrado com sucesso!")

# Parte 3: Desenvolvimento das funções para consultar clientes, serviços e agendamentos
# Função para consultar clientes
def consultar_clientes():
  if not clientes:
      print("Nenhum cliente cadastrado.")
  else:
      print("\nLista de Clientes:")
      for cpf, cliente in clientes.items():
          print(f'CPF: {cpf}, Nome: {cliente["Nome"]}, Endereço: {cliente["Endereço"]}, Telefone: {cliente["Telefone"]}, Email: {cliente["Email"]}')

# Função para consultar serviços
def consultar_servicos():
  if not servicos:
      print("Nenhum serviço cadastrado.")
  else:
      print("\nLista de Serviços:")
      for codigo, servico in servicos.items():
          print(f'Código: {codigo}, Descrição: {servico["Descrição"]}, Preço: {servico["Preço"]}')

# Função para consultar agendamentos
def consultar_agendamentos():
  if not agendamentos:
      print("Nenhum agendamento cadastrado.")
  else:
      print("\nLista de Agendamentos:")
      for codigo, agendamento in agendamentos.items():
          cliente = clientes.get(agendamento["CPF Cliente"], {"Nome": "Cliente não encontrado"})
          servico = servicos.get(agendamento["Código Serviço"], {"Descrição": "Serviço não encontrado"})
          print(f'Código: {codigo}, Cliente: {cliente["Nome"]}, Serviço: {servico["Descrição"]}, Data: {agendamento["Data"]}, Hora: {agendamento["Hora"]}')

# Parte 4: Desenvolvimento das funções para atualizar clientes, serviços e agendamentos
# Função para atualizar clientes
def atualizar_cliente():
  cpf = input("Digite o CPF do cliente a ser atualizado: ")
  if cpf not in clientes:
      print(f'Cliente com CPF "{cpf}" não encontrado.')
      return
  nome = input("Nome atualizado: ")
  endereco = input("Endereço atualizado: ")
  telefone = input("Telefone atualizado: ")
  email = input("Email atualizado: ")
  clientes[cpf] = {"Nome": nome, "Endereço": endereco, "Telefone": telefone, "Email": email}
  print(f'Cliente "{cpf}" atualizado com sucesso!')

# Função para atualizar serviços
def atualizar_servico():
  codigo = input("Digite o código do serviço a ser atualizado: ")
  if codigo not in servicos:
      print(f'Serviço com código "{codigo}" não encontrado.')
      return
  descricao = input("Descrição atualizada: ")
  preco = input("Preço atualizado: ")
  servicos[codigo] = {"Descrição": descricao, "Preço": preco}
  print(f'Serviço "{codigo}" atualizado com sucesso!')

# Função para atualizar agendamentos
def atualizar_agendamento():
    codigo = input("Digite o código do agendamento a ser atualizado: ")
    if codigo not in agendamentos:
        print(f'Agendamento com código "{codigo}" não encontrado.')
        return
    cod_cliente = input("CPF atualizado do Cliente: ")
    if cod_cliente not in clientes:
        print(f'Erro: Cliente com CPF "{cod_cliente}" não existe.')
        return
    cod_servico = input("Código atualizado do Serviço: ")
    if cod_servico not in servicos:
        print(f'Erro: Serviço com código "{cod_servico}" não existe.')
        return
    data = input("Data atualizada (dd/mm/yyyy): ")
    hora = input("Hora atualizada (HH:MM): ")
    agendamentos[codigo] = {"CPF Cliente": cod_cliente, "Código Serviço": cod_servico, "Data": data, "Hora": hora}
    print(f'Agendamento "{codigo}" atualizado com sucesso!')

# Parte 5: Desenvolvimento das funções para remoção de clientes, serviços e agendamentos
# Função para remover clientes
def remover_cliente():
    cpf = input("Digite o CPF do cliente a ser removido: ")
    if cpf in clientes:
        del clientes[cpf]
        print(f'Cliente "{cpf}" removido com sucesso!')
    else:
        print(f'Cliente com CPF "{cpf}" não encontrado.')

# Função para remover serviços
def remover_servico():
    codigo = input("Digite o código do serviço a ser removido: ")
    if codigo in servicos:
        del servicos[codigo]
        print(f'Serviço "{codigo}" removido com sucesso!')
    else:
        print(f'Serviço com código "{codigo}" não encontrado.')

# Função para remover agendamentos
def remover_agendamento():
    codigo = input("Digite o código do agendamento a ser removido: ")
    if codigo in agendamentos:
        del agendamentos[codigo]
        print(f'Agendamento "{codigo}" removido com sucesso!')
    else:
        print(f'Agendamento com código "{codigo}" não encontrado.')

# Programa Principal
# Execução do aplicativo
if __name__ == "__main__":
    main()
```
